```python
def normalize(result):
    result = np.array(result)
    norm = []
    for i in result:
#         norm.append((i - result.min() + 10)/ (result.max() - result.min()))
        norm.append(i)
    return norm
```

```python
def get_result(feature_type, index, result):
    recall = []
    precision = []
    accuracy = []
    F1 = []
    for model in result:
        recall.append(model[index][0])
        precision.append(model[index][1])
        accuracy.append(model[index][2])
        F1.append(model[index][3])
    recall = normalize(recall)
    precision = normalize(precision)
    accuracy = normalize(accuracy)
    F1 = normalize(F1)
    return [recall, precision, accuracy, F1]


def show_result(result):
    location = 221
    plt.figure(figsize=(20, 10))
    for i in range(4):
        plt.subplot(location + i)
        plt.bar(x, result[0][i], bar_width, color='#95a2ff', label='A')
        plt.bar(x + bar_width, result[1][i], bar_width, color='#fa8080', label='B')
        plt.bar(x + 2 * bar_width, result[2][i], bar_width, color='#ffc076', label='C')
        plt.title(name[i], fontsize=40)
        plt.legend(fontsize=20, loc='best')
        plt.xticks(x + bar_width, tick_label, fontsize=30)
        plt.yticks(fontsize=30)
        plt.ylim(ymin = 70)
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
```

```python
tick_label = ['ADT', 'RF', 'DEC', 'NN']
name = ['recall', 'precision', 'accuracy', 'F1']
x = np.arange(4)
bar_width = 0.3
A = get_result('A', 0, result)
B = get_result('B', 1, result)
C = get_result('C', 2, result)
show_result([A, B, C])
```
